###### Defining function

```python
def remove_sign(x, sign):
    '''Replaces given sign in applied dataframe'''
    if type(x) is str:
        x = float(x.replace(sign,'').replace(',',''))
    return x
```

_This function wich will help me to clear prices, because they have $ sign and have object type_

###### Choosing the right data types

```python
calendar.price = calendar.price.apply(remove_sign, sign='$')
```

_First, lest's remove $ from calendar.price_

```python
calendar.date = calendar.date.apply(lambda x: datetime.strptime(str(x), '%Y-%m-%d'))
```

_Second, let's change calendar.date from string to datetime_

## How many offers in the marketplace and what is the average price during the year?

_Grouping available properties by day. Calculating the count of listings and the average price of them._

```python
available_by_day = calendar[calendar['available'] == 't'].groupby('date', as_index = False).agg({'listing_id': 'count', 
                                                                                                 'price': 'mean'})
available_by_day = available_by_day.rename({'listing_id': 'listing_count', 'price': 'average_price'}, axis='columns')
available_by_day.head(5)
```

```python
f, ax = plt.subplots(figsize=(20, 10))

plt1 = sns.lineplot(x=available_by_day.date, y='listing_count', color='r', data=available_by_day, legend=False)
for tl in ax.get_yticklabels():
    tl.set_color('r')
    
ax2 = ax.twinx()
plt2 = sns.lineplot(x=available_by_day.date, y='average_price', color='b', data=available_by_day, legend=False)
for t2 in ax2.get_yticklabels():
    t2.set_color('b')
    
ax.set_xlabel('Date')
ax.set_ylabel('Listings count')
ax2.set_ylabel('Average price')

plt.show()
```

_We can see, that availability is not stable all year round and price grows evenly until mid-summer, then it slightly decreases by the end of the year. This may be due to the popularity of the city for tourists in the summer._

## What is the price depending on the neighborhood?

_Reading sharp file with GeoPandas and plotting the map with additional data._

```python
seattle_map = gpd.read_file('Community_Reporting_Areas.shp')
```

```python
fig, ax = plt.subplots(figsize = (20, 20))
seattle_map.plot(ax=ax, color='grey', alpha = 0.3)

ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
c = listings_filtered['price']
plt.scatter(listings_filtered['longitude'], 
            listings_filtered['latitude'], 
            c=c,
            cmap='OrRd',
            s=listings_filtered['price']*0.3, 
            alpha=0.8)
cbar = plt.colorbar()
cbar.set_label('Price')
plt.show()
```

_As expected a large number of homestays are at the center of the city, but they have a higher average price. Let’s take a closer look at the distribution of prices by neighborhoods._

_Boxplots and barcharts_

```python
f, ax = plt.subplots(figsize=(20, 10))
sns.boxplot(x = 'neighbourhood_cleansed', y = 'price',  data = listings_filtered)
xt = plt.xticks(rotation=90)

ax.set_xlabel('Neighbourhoods')
ax.set_ylabel('Price')

plt.show()
```

_We can see a lot of neighborhoods and some of them have a large variance. Below is a graph with the number of listings divided by neighborhoods._

```python
by_neighbourhood = listings_filtered.groupby('neighbourhood_cleansed', as_index=False).agg({'id': 'count', 
                                                                                            'price': 'mean'})
by_neighbourhood = by_neighbourhood.rename({'id': 'listings_count', 
                                            'price': 'average_price'}, axis='columns')
by_neighbourhood = by_neighbourhood.sort_values(['listings_count'], ascending=False).reset_index(drop=True)
by_neighbourhood
```

```python
fig, ax = plt.subplots(figsize = (20, 30))
ax = sns.barplot(x='listings_count', y='neighbourhood_cleansed', palette='Reds_d', data=by_neighbourhood)
ax.set_xlabel('Listings count')
ax.set_ylabel('Neighbourhoods')
plt.show()
```

_The given statistic is not very informative. So, let’s use groups of neighborhoods form Airbnb and do the steps above._

```python
f, ax = plt.subplots(figsize=(20, 10))
sns.boxplot(x = 'neighbourhood_group_cleansed', y = 'price',  data = listings_filtered)
xt = plt.xticks(rotation=90)

ax.set_xlabel('Grouped neighbourhoods')
ax.set_ylabel('Price')

plt.show()
```

_We can see, that Queen Anne, Downtown and Magnolia have a higher median than others, whereas Lake City, Delridge, University District has more attractive prices. Also, all of the grouped neighborhoods have some outliers, but most are located in more expensive areas._

```python
by_neighbourhood_grouped = listings_filtered.groupby('neighbourhood_group_cleansed', as_index=False).agg({'id': 'count', 
                                                                                                          'price': 'mean'})
by_neighbourhood_grouped = by_neighbourhood_grouped.rename({'id': 'listings_count', 
                                                            'price': 'average_price'}, axis='columns')
by_neighbourhood_grouped = by_neighbourhood_grouped.sort_values(['listings_count'], ascending=False).reset_index(drop=True)
by_neighbourhood_grouped
```

```python
fig, ax = plt.subplots(figsize = (20, 10))
ax = sns.barplot(x='listings_count', y='neighbourhood_group_cleansed', palette='Reds_d', data=by_neighbourhood_grouped)
ax.set_xlabel('Listings count')
ax.set_ylabel('Grouped neighbourhoods')
plt.show()
```

_Also, we can notice that most of the listings are located in Capitol Hill, Downtown, Central Area and Queen Anne where prices are higher. So you may encounter difficulties when searching in other districts._

## What parameters can affect the price?

_Potting heatmaps_

_Grouped neighbourhoods & Bedrooms with Average price_

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(listings_filtered.groupby(['neighbourhood_group_cleansed', 'bedrooms'],).price.mean().unstack(),
            annot=True, 
            fmt=".0f", 
            cmap = 'OrRd',
            cbar_kws={'label': 'Average price'}
           )

ax.set_xlabel('Bedrooms')
ax.set_ylabel('Grouped neighbourhoods')

plt.show()
```

_Here we can see, that listings are more expensive in Capitol Hill, Cascade, Queen Anne and in the Downtown. We saw similar results above. But if you are looking for a small homestay with affordable prices you might be interested in such neighborhoods like Beacon Hill, Lake City or Rainier Valley._

_Property types & Bedrooms with Average price_

```python
f, ax = plt.subplots(figsize=(10, 10))
sns.heatmap(listings_filtered.groupby(['property_type', 'bedrooms'],).price.mean().unstack(),
            annot=True, 
            fmt=".0f", 
            cmap = 'OrRd',
            cbar_kws={'label': 'Average price'}
           )

ax.set_xlabel('Bedrooms')
ax.set_ylabel('Property type')

plt.show()
```

_Such properties like boats, houses also have a higher average price than others. But if you don’t want to spend too much money, you can rent a room in the dorm or tent and treehouse :)_

_Grouped neighbourhoods & Property types with Average price_

```python
f, ax = plt.subplots(figsize=(15, 10))
sns.heatmap(listings_filtered.groupby(['neighbourhood_group_cleansed', 'property_type'],).price.mean().unstack(),
            annot=True, 
            fmt=".0f", 
            cmap = 'OrRd',
            cbar_kws={'label': 'Average price'}
           )

ax.set_xlabel('Property type')
ax.set_ylabel('Grouped neighbourhoods')

plt.show()
```

_Here we can see, that each neighborhood has its types of properties. And if you are looking for the best option wich close to the city center and low average price, there are dorms Downtown with the average price of 40$._
